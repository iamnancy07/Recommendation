```python
from xlrd import open_workbook

wb = open_workbook('LinkedPages.xlsx')

for s in wb.sheets():
    
    values = []
    
    for row in range(s.nrows):
        
        for col in range(s.ncols):
            
            values.append(s.cell(row,col).value)
```

```python
import re

#inputList = ["x-y-z", "x-y", "x-m", "y-z-m","x-y-n","y-o-m","z-n-o"]

outputList = []

for count, inputStr in enumerate(values):
    
    
    inputStr = re.sub('\s+', ' ', inputStr)
    

    
    if inputStr.find("---") > 1:
        
        inputStrToList = inputStr.split("---")
        #print inputStrToList
     
        if len(inputStrToList) <= 2:
            
            outputList.append
            outputList.append(inputStrToList)
            

        else:
            
            for i in range(len(inputStrToList) - 1):
                

                subList = [] 
                
                subList.append            
                subList.append(inputStrToList[i])
                subList.append(inputStrToList[i+1])
                 
                outputList.append(subList)
```

```python
outputList[1:3]
```

```python
import pandas as pd

test = pd.DataFrame(outputList)

test.columns = ['Origin','Linked']    
test = test[test.Origin != '']     
test["Dummy"] = 1
test1 = test.groupby('Origin', as_index = False, sort = False)['Dummy'].sum()
test2 = test.groupby(['Origin','Linked'], as_index = False, sort = False)['Dummy'].sum()   
    
test3 = pd.merge(test1, test2, how = 'inner', left_on = 'Origin', right_on = 'Origin')       
        
test3['Percent'] = test3.Dummy_y / test3.Dummy_x

test4 = test3.sort(['Dummy_x', 'Origin', 'Dummy_y'], ascending = [False, True, False])
    
test5 = test4.groupby('Origin').head(3).reset_index(drop = True)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(121)

numBins = 50
ax1.hist(test5.Percent,numBins,color='green',alpha=0.8)

ax2 = fig.add_subplot(122)

ax2.hist(test5.Dummy_x,numBins,color='red',alpha=0.8)


plt.show()
plt.close()
```
